# Dilution Refrigerator Data

Use the first set of data at 3 kelvin from the "6x6cal1" data set to calibrate the raw data from a dilution refrigerator measurement. Standards in this cooldown are in "F" configuration, which is with the flat short on port 6, the open on port 5, and offset shorts of increasing length on ports 1, 2, 3, and 4.

```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
```

```python
# read in raw data
rawdata_base = rf.read_all('dilution_refrigerator_data/base')
rawdata_100mk = rf.read_all('dilution_refrigerator_data/100mk')
rawdata_1k = rf.read_all('dilution_refrigerator_data/1k')
rawdata_3k = rf.read_all('dilution_refrigerator_data/3k')
rawdata_roomtemp = rf.read_all('dilution_refrigerator_data/roomtemp')
# define sets of mesaured data from ecal for ecal calibration at four cold temperatures
my_measured_base = [\
        rawdata_base['ecal_short_base'],
        rawdata_base['ecal_open_base'],
        rawdata_base['ecal_load_base'],
        ]
my_measured_100mk = [\
        rawdata_100mk['ecal_short_100mk'],
        rawdata_100mk['ecal_open_100mk'],
        rawdata_100mk['ecal_load_100mk'],
        ]
my_measured_1k = [\
        rawdata_1k['ecal_short_1k'],
        rawdata_1k['ecal_open_1k'],
        rawdata_1k['ecal_load_1k'],
        ]
my_measured_3k = [\
        rawdata_3k['ecal_short_3k'],
        rawdata_3k['ecal_open_3k'],
        rawdata_3k['ecal_load_3k'],
        ]
# read in data from tier 2 scikit-rf calibration data files
tier2_3k1_cal_files_scikitrf = rf.read_all('tier2_scikitrf_caldata/tier2_3k1')

# define sets of ideals from tier 2 at each port from the first 6x6cal data set at 3k
my_ideals_scikitrf_port1_3k1 = [\
        tier2_3k1_cal_files_scikitrf['port1_short_tier2'],
        tier2_3k1_cal_files_scikitrf['port1_open_tier2'],
        tier2_3k1_cal_files_scikitrf['port1_load_tier2'],
        ]
my_ideals_scikitrf_port2_3k1 = [\
        tier2_3k1_cal_files_scikitrf['port2_short_tier2'],
        tier2_3k1_cal_files_scikitrf['port2_open_tier2'],
        tier2_3k1_cal_files_scikitrf['port2_load_tier2'],
        ]
my_ideals_scikitrf_port3_3k1 = [\
        tier2_3k1_cal_files_scikitrf['port3_short_tier2'],
        tier2_3k1_cal_files_scikitrf['port3_open_tier2'],
        tier2_3k1_cal_files_scikitrf['port3_load_tier2'],
        ]
my_ideals_scikitrf_port4_3k1 = [\
        tier2_3k1_cal_files_scikitrf['port4_short_tier2'],
        tier2_3k1_cal_files_scikitrf['port4_open_tier2'],
        tier2_3k1_cal_files_scikitrf['port4_load_tier2'],
        ]
my_ideals_scikitrf_port5_3k1 = [\
        tier2_3k1_cal_files_scikitrf['port5_short_tier2'],
        tier2_3k1_cal_files_scikitrf['port5_open_tier2'],
        tier2_3k1_cal_files_scikitrf['port5_load_tier2'],
        ]
my_ideals_scikitrf_port6_3k1 = [\
        tier2_3k1_cal_files_scikitrf['port6_short_tier2'],
        tier2_3k1_cal_files_scikitrf['port6_open_tier2'],
        tier2_3k1_cal_files_scikitrf['port6_load_tier2'],
        ]

```

```python
# define cal standards sets at each port at each (cold) temperature
cal_scikitrf_base_port1 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port1_3k1,
        measured = my_measured_base,
        )

cal_scikitrf_base_port2 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port2_3k1,
        measured = my_measured_base,
        )

cal_scikitrf_base_port3 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port3_3k1,
        measured = my_measured_base,
        )

cal_scikitrf_base_port4 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port4_3k1,
        measured = my_measured_base,
        )

cal_scikitrf_base_port5 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port5_3k1,
        measured = my_measured_base,
        )

cal_scikitrf_base_port6 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port6_3k1,
        measured = my_measured_base,
        )

cal_scikitrf_100mk_port1 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port1_3k1,
        measured = my_measured_100mk,
        )

cal_scikitrf_100mk_port2 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port2_3k1,
        measured = my_measured_100mk,
        )

cal_scikitrf_100mk_port3 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port3_3k1,
        measured = my_measured_100mk,
        )

cal_scikitrf_100mk_port4 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port4_3k1,
        measured = my_measured_100mk,
        )

cal_scikitrf_100mk_port5 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port5_3k1,
        measured = my_measured_100mk,
        )

cal_scikitrf_100mk_port6 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port6_3k1,
        measured = my_measured_100mk,
        )
cal_scikitrf_1k_port1 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port1_3k1,
        measured = my_measured_1k,
        )

cal_scikitrf_1k_port2 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port2_3k1,
        measured = my_measured_1k,
        )

cal_scikitrf_1k_port3 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port3_3k1,
        measured = my_measured_1k,
        )

cal_scikitrf_1k_port4 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port4_3k1,
        measured = my_measured_1k,
        )

cal_scikitrf_1k_port5 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port5_3k1,
        measured = my_measured_1k,
        )

cal_scikitrf_1k_port6 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port6_3k1,
        measured = my_measured_1k,
        )
cal_scikitrf_3k_port1 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port1_3k1,
        measured = my_measured_3k,
        )

cal_scikitrf_3k_port2 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port2_3k1,
        measured = my_measured_3k,
        )

cal_scikitrf_3k_port3 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port3_3k1,
        measured = my_measured_3k,
        )

cal_scikitrf_3k_port4 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port4_3k1,
        measured = my_measured_3k,
        )

cal_scikitrf_3k_port5 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port5_3k1,
        measured = my_measured_3k,
        )

cal_scikitrf_3k_port6 = rf.OnePort(\
        ideals = my_ideals_scikitrf_port6_3k1,
        measured = my_measured_3k,
        )

```

```python
# run all the calibrations at each port at each temperature
cal_scikitrf_base_port1.run()
cal_scikitrf_base_port2.run()
cal_scikitrf_base_port3.run()
cal_scikitrf_base_port4.run()
cal_scikitrf_base_port5.run()
cal_scikitrf_base_port6.run()
cal_scikitrf_100mk_port1.run()
cal_scikitrf_100mk_port2.run()
cal_scikitrf_100mk_port3.run()
cal_scikitrf_100mk_port4.run()
cal_scikitrf_100mk_port5.run()
cal_scikitrf_100mk_port6.run()
cal_scikitrf_1k_port1.run()
cal_scikitrf_1k_port2.run()
cal_scikitrf_1k_port3.run()
cal_scikitrf_1k_port4.run()
cal_scikitrf_1k_port5.run()
cal_scikitrf_1k_port6.run()
cal_scikitrf_3k_port1.run()
cal_scikitrf_3k_port2.run()
cal_scikitrf_3k_port3.run()
cal_scikitrf_3k_port4.run()
cal_scikitrf_3k_port5.run()
cal_scikitrf_3k_port6.run()
```

```python
# apply all calibration corrections to all raw data
port1_base_scikitrf_corrected = cal_scikitrf_base_port1.apply_cal(rawdata_base['port1_base'])
port2_base_scikitrf_corrected = cal_scikitrf_base_port2.apply_cal(rawdata_base['port2_base'])
port3_base_scikitrf_corrected = cal_scikitrf_base_port3.apply_cal(rawdata_base['port3_base'])
port4_base_scikitrf_corrected = cal_scikitrf_base_port4.apply_cal(rawdata_base['port4_base'])
port5_base_scikitrf_corrected = cal_scikitrf_base_port5.apply_cal(rawdata_base['port5_base'])
port6_base_scikitrf_corrected = cal_scikitrf_base_port6.apply_cal(rawdata_base['port6_base'])
port1_100mk_scikitrf_corrected = cal_scikitrf_100mk_port1.apply_cal(rawdata_100mk['port1_100mk'])
port2_100mk_scikitrf_corrected = cal_scikitrf_100mk_port2.apply_cal(rawdata_100mk['port2_100mk'])
port3_100mk_scikitrf_corrected = cal_scikitrf_100mk_port3.apply_cal(rawdata_100mk['port3_100mk'])
port4_100mk_scikitrf_corrected = cal_scikitrf_100mk_port4.apply_cal(rawdata_100mk['port4_100mk'])
port5_100mk_scikitrf_corrected = cal_scikitrf_100mk_port5.apply_cal(rawdata_100mk['port5_100mk'])
port6_100mk_scikitrf_corrected = cal_scikitrf_100mk_port6.apply_cal(rawdata_100mk['port6_100mk'])
port1_1k_scikitrf_corrected = cal_scikitrf_1k_port1.apply_cal(rawdata_1k['port1_1k'])
port2_1k_scikitrf_corrected = cal_scikitrf_1k_port2.apply_cal(rawdata_1k['port2_1k'])
port3_1k_scikitrf_corrected = cal_scikitrf_1k_port3.apply_cal(rawdata_1k['port3_1k'])
port4_1k_scikitrf_corrected = cal_scikitrf_1k_port4.apply_cal(rawdata_1k['port4_1k'])
port5_1k_scikitrf_corrected = cal_scikitrf_1k_port5.apply_cal(rawdata_1k['port5_1k'])
port6_1k_scikitrf_corrected = cal_scikitrf_1k_port6.apply_cal(rawdata_1k['port6_1k'])
port1_3k_scikitrf_corrected = cal_scikitrf_3k_port1.apply_cal(rawdata_3k['port1_3k'])
port2_3k_scikitrf_corrected = cal_scikitrf_3k_port2.apply_cal(rawdata_3k['port2_3k'])
port3_3k_scikitrf_corrected = cal_scikitrf_3k_port3.apply_cal(rawdata_3k['port3_3k'])
port4_3k_scikitrf_corrected = cal_scikitrf_3k_port4.apply_cal(rawdata_3k['port4_3k'])
port5_3k_scikitrf_corrected = cal_scikitrf_3k_port5.apply_cal(rawdata_3k['port5_3k'])
port6_3k_scikitrf_corrected = cal_scikitrf_3k_port6.apply_cal(rawdata_3k['port6_3k'])
```
