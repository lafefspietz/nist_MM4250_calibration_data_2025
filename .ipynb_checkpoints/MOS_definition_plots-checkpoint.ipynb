```python
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
```

```python
mosdef = rf.read_all('MOS_MUF_model_construction/MOS_definitions_touchstone') # recalculated with final value of gold conducivity
```

```python
mosdef
```

```python
fghz = mosdef['MOS1'].f/1e9
plt.plot(fghz,mosdef['MOS1'].s_db[:,0,0],color="red")
plt.plot(fghz,mosdef['MOS2_warm'].s_db[:,0,0],color="orange")
plt.plot(fghz,mosdef['MOS3_warm'].s_db[:,0,0],color="yellow")
plt.plot(fghz,mosdef['MOS4_warm'].s_db[:,0,0],color="green")
plt.plot(fghz,mosdef['MOS5_warm'].s_db[:,0,0],color="blue")
plt.plot(fghz,mosdef['MOS6_warm'].s_db[:,0,0],color="purple")
plt.grid()
plt.xlim(0.001,12)
plt.ylim(-0.15,0)
plt.title('Calculated MOS Standards, room temperature')
plt.xlabel('Frequency [GHz]')
plt.ylabel('$S_{11}$ [dB]')
plt.legend(('flat short',"5 mm offset short","10.8 mm offset short","18 mm offset short","30 mm offset short","open"))
```

```python
fghz = mosdef['MOS1'].f/1e9
plt.plot(fghz,mosdef['MOS1'].s_db[:,0,0],color="red")
plt.plot(fghz,mosdef['MOS2_cold'].s_db[:,0,0],color="orange")
plt.plot(fghz,mosdef['MOS3_cold'].s_db[:,0,0],color="yellow")
plt.plot(fghz,mosdef['MOS4_cold'].s_db[:,0,0],color="green")
plt.plot(fghz,mosdef['MOS5_cold'].s_db[:,0,0],color="blue")
plt.plot(fghz,mosdef['MOS6_cold'].s_db[:,0,0],color="purple")
plt.grid()
plt.xlim(0.001,12)
plt.ylim(-0.15,0)

plt.title('Calculated MOS Standards, 4 kelvin')
plt.xlabel('Frequency [GHz]')
plt.ylabel('$S_{11}$ [dB]')
plt.legend(('flat short',"5 mm offset short","10.8 mm offset short","18 mm offset short","30 mm offset short","open"))
```
