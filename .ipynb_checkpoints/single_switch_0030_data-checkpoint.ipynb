```python
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
```

```python
all_data = rf.read_all('single_switch_0030_data')
```

```python
fghz = all_data['port1_295k'].f/1e9 # create a numpy array with frequency points in GHz
```

```python
plt.plot(fghz,all_data['port1_295k'].s_db[:,1,0],color='red')
plt.plot(fghz,all_data['port1_3k'].s_db[:,1,0],color='blue')
plt.grid()
plt.xlim(0.1,15)
plt.ylim(-30,0)
plt.title('Raw Switch Port 1 Through Data No Subtraction')
plt.xlabel('Frequency [GHz]')
plt.ylabel('$S_{21}$ [dB]')
plt.legend(('T = 295 K','T = 3 K'))
```

```python
plt.plot(fghz,all_data['thru_295k'].s_db[:,1,0],color='red')
plt.plot(fghz,all_data['thru_3k'].s_db[:,1,0],color='blue')
plt.grid()
plt.xlim(0.1,15)
plt.ylabel('$S_{21}$ [dB]')
plt.ylim(-30,0)
plt.xlabel('Frequency [GHz]')
plt.title('Through, No Switch')
plt.legend(('T = 295 K','T = 3 K'))
```

```python
plt.plot(fghz,all_data['port1_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0],color='red')
plt.plot(fghz,all_data['port1_3k'].s_db[:,1,0] - all_data['thru_3k'].s_db[:,1,0],color='blue')
plt.grid()
plt.xlim(0.1,15)
plt.ylim(-10,0)
plt.ylabel('$S_{21}$ [dB]')
plt.xlabel('Frequency [GHz]')
plt.title('Switch Port 1 Insertion Loss')
plt.legend(('T = 295 K','T = 3 K'))
```

```python
plt.plot(fghz,all_data['port1_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0] - all_data['port1_3k'].s_db[:,1,0] + all_data['thru_3k'].s_db[:,1,0],color='purple')
plt.grid()
plt.xlim(0.1,15)
plt.ylim(-5,0)
plt.ylabel('$S_{21}$ [dB]')
plt.xlabel('Frequency [GHz]')
plt.title('295 K - 3 K Insertion Loss')
plt.legend(('T = 295 K','T = 3 K'))
```

```python
plt.plot(fghz,all_data['port2_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0],color='red')
plt.plot(fghz,all_data['port3_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0],color='orange')
plt.plot(fghz,all_data['port4_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0],color='yellow')
plt.plot(fghz,all_data['port5_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0],color='green')
plt.plot(fghz,all_data['port6_295k'].s_db[:,1,0] - all_data['thru_295k'].s_db[:,1,0],color='blue')

plt.grid()
plt.xlim(0.1,15)
plt.title('Isolation 295 K')
plt.ylim(-100,0)
plt.xlabel('Frequency [GHz]')
plt.legend(('Switch State Port2','Switch State Port3','Switch State Port4','Switch State Port5','Switch State Port6'))
plt.ylabel('Port 1 Through Insertion Loss [dB]')
```

```python
plt.plot(fghz,all_data['port2_3k'].s_db[:,1,0] - all_data['thru_3k'].s_db[:,1,0],color='red')
plt.plot(fghz,all_data['port3_3k'].s_db[:,1,0] - all_data['thru_3k'].s_db[:,1,0],color='orange')
plt.plot(fghz,all_data['port4_3k'].s_db[:,1,0] - all_data['thru_3k'].s_db[:,1,0],color='yellow')
plt.plot(fghz,all_data['port5_3k'].s_db[:,1,0] - all_data['thru_3k'].s_db[:,1,0],color='green')
plt.plot(fghz,all_data['port6_3k'].s_db[:,1,0] - all_data['thru_3k'].s_db[:,1,0],color='blue')

plt.grid()
plt.xlim(0.1,15)
plt.title('Isolation At 3 K')
plt.ylim(-80,0)
plt.xlabel('Frequency [GHz]')
plt.ylabel('Port 1 Through Insertion Loss [dB]')
plt.legend(('Switch State Port2','Switch State Port3','Switch State Port4','Switch State Port5','Switch State Port6'))
```

```python
plt.plot(fghz,all_data['port1_295k'].s_db[:,0,0],color='red')
plt.plot(fghz,all_data['port1_3k'].s_db[:,0,0],color='blue')
plt.grid()
plt.xlim(0.1,15)
plt.ylim(-40,0)
plt.ylabel('$S_{11}$ [dB]')
plt.title('Raw Uncorrected $S_{11}$ of Input Connected to Port 1 Thru')
plt.xlabel('Frequency [GHz]')
plt.legend(('T = 295 K','T = 3 K'))
```

```python
plt.plot(fghz,all_data['shorts_295k'].s_db[:,0,0],color='red')
plt.plot(fghz,all_data['shorts_3k'].s_db[:,0,0],color='blue')
plt.grid()
plt.xlim(0.1,15)
plt.ylim(-40,0)
plt.ylabel('$S_{11}$ [dB]')
plt.title('Shorts Raw Data')
plt.xlabel('Frequency [GHz]')
plt.legend(('T = 295 K','T = 3 K'))
```

```python
plt.plot(fghz,all_data['port1_295k'].s_db[:,0,0] - all_data['shorts_295k'].s_db[:,0,0],color='red')
plt.plot(fghz,all_data['port1_3k'].s_db[:,0,0] - all_data['shorts_3k'].s_db[:,0,0],color='blue')
plt.grid()
plt.xlim(0.1,15)
plt.ylim(-40,0)
plt.ylabel('$S_{11}$ [dB]')
plt.title('Insertion Loss with Shorts Traces Subtracted')
plt.xlabel('Frequency [GHz]')
plt.legend(('T = 295 K','T = 3 K'))
```

```python

for tracename in all_data:
    if 'port1_3k' in tracename:
        if not 'afterswitches' in tracename:
            plt.plot(fghz,all_data[tracename].s_db[:,1,0] - all_data['port1_3k'].s_db[:,1,0],color='black')
for tracename in all_data:
    if 'port1_3k' in tracename:
        if 'afterswitches' in tracename:
            plt.plot(fghz,all_data[tracename].s_db[:,1,0] - all_data['port1_3k'].s_db[:,1,0],color='purple')

plt.grid()
plt.xlim(0.1,15)
plt.ylim(-0.1,0.1)
plt.ylabel('$S_{11}$ [dB]')
plt.xlabel('Frequency [GHz]')
plt.title('Insertion Loss Difference From First Trace In Set')
```
